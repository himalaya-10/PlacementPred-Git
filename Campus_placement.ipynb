```python
#using pd, we are reading dataset from csv file and storing the dataset in pdata variable.
pdata=pd.read_csv('allpgrmplacement.csv')
```

```python
#info() to specfically get datatypes of features in dataset (used for cleaning preprocess)
pdata.info()
```

```python
#to get initial instances with better dataframe.
pdata.head()
```

```python
#to know no. of rows and columns in dataset.
pdata.shape
```

```python
#converting status placed:1 and notplaced:0 in column.

def convertPlacedOrNot(value):
    if value=='Placed' :
        return 1;
    elif value=='Not Placed' :
        return 0;
    
pdata['status']=pdata['status'].apply(convertPlacedOrNot)
```

```python
#as the data is already clean we will now convert categorical features into numerical features using onehotencoder.
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
xd=pdata.drop(columns=['sl_no','salary'])
```

```python
xd
```

```python
#fiting the feature need to be converted.
ohe.fit(xd[['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation']])
```

```python
#seeing what unique values does the feautures consist of with datatypes.
ohe.categories_
```

```python
#to convert the colums of x we import make column transformer.
from sklearn.compose import make_column_transformer
```

```python
#we assign what method is used (ohe) and the columns to be converted to make_column_transformer.
#(case: remainder='passthrough' if a column not mentioned in column categories[] which is already in numeric value to passthrough as it is.)
column_transfer=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation']),remainder='passthrough')
```

```python
column_transfer
```

```python
#fitting the transformed columns to x dataset with fit_transform method.
transformeddata=column_transfer.fit_transform(xd)
```

```python
transformeddata
```

```python
#making a proper dataframe and using get_features_name_out from x.columns to automatically assign column name to dataframe.
transformed_df = pd.DataFrame(transformeddata, columns=column_transfer.get_feature_names_out(xd.columns))
```

```python
transformed_df
```

```python
#renaming columns for better readability.
transformed_df.rename(columns={'onehotencoder__gender_F':'F','onehotencoder__gender_M':'M','onehotencoder__ssc_b_Central':'boardofsscCentral','onehotencoder__ssc_b_Others':'boardofsscOther','onehotencoder__hsc_b_Central':'boardofhscCentral','onehotencoder__hsc_b_Others':'boardofhscOther','onehotencoder__hsc_s_Arts':'specArts','onehotencoder__hsc_s_Commerce':'specCommerce','onehotencoder__hsc_s_Science':'specScience','onehotencoder__degree_t_Comm&Mgmt':'degcomm&mgmt','onehotencoder__degree_t_Others ':'degreeOthers','onehotencoder__degree_t_Sci&Tech':'degreesci&tech','onehotencoder__workex_No':'expN','onehotencoder__workex_Yes':'expY','onehotencoder__specialisation_Mkt&Fin':'MspecMkt&Fin','onehotencoder__specialisation_Mkt&HR':'MspecMkt&HR','remainder__ssc_p':'sscp','remainder__hsc_p':'hscp','remainder__degree_p':'degp','remainder__etest_p':'etestp','remainder__mba_p':'mbap','remainder__status':'status'},inplace=True)
```

```python
transformed_df
```

```python
transformed_df.info()
```

```python
## as the datatype of all the features are float now cleaning has been done. Lets proceed with processing.
cleaned_data=transformed_df
#assigning x and y.
x=cleaned_data.drop(columns=['status'])
y=cleaned_data['status']
```

```python
#importing train test split
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
#importing classification regression models
from sklearn.linear_model import LogisticRegression
Lor=LogisticRegression(solver='lbfgs', max_iter=1000)#to increase max_iteration (warning)
from sklearn.tree import DecisionTreeClassifier
Dt=DecisionTreeClassifier()
from sklearn.ensemble import RandomForestClassifier
Rf=RandomForestClassifier()

```

```python
#fitting classification model on dataset.
Lor.fit(xtrain,ytrain)
```

```python
Dt.fit(xtrain,ytrain)
```

```python
Rf.fit(xtrain,ytrain)
```

```python
#library to check the classification is how much accurate.
from sklearn.metrics import confusion_matrix
```

```python
ypred_lor=Lor.predict(xtest)
ypred_dt=Dt.predict(xtest)
ypred_rf=Rf.predict(xtest)
```

```python
ypred_lor
```

```python
ypred_dt
```

```python
ypred_rf
```

```python
confusion_matrix(ytest,ypred_lor)
```

```python
confusion_matrix(ytest,ypred_dt)
```

```python
confusion_matrix(ytest,ypred_rf)
```

```python
#classification has been done,now we can choose any classification model among them but for better accuracy I will choose random_forest
#making the model file for prediction
import pickle
pickle.dump(Rf, open('placementmodel.pkl','wb'))
```

```python
#checking the model
model=pickle.load(open('placementmodel.pkl','rb'))
ypred_rf2=model.predict(xtest)
```
